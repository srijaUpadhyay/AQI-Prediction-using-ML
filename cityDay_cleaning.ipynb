```python
# --- AQI Calculation Functions and Breakpoints (ESSENTIAL) ---
# You MUST have these defined in your script before calling calculate_aqi_for_dataframe_daily

# 1. CPCB AQI Breakpoints
cpcb_breakpoints = {
    'PM2.5': [(0, 30, 0, 50), (31, 60, 51, 100), (61, 90, 101, 200), (91, 120, 201, 300), (121, 250, 301, 400), (251, np.inf, 401, 500)],
    'PM10': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 250, 101, 200), (251, 350, 201, 300), (351, 430, 301, 400), (431, np.inf, 401, 500)],
    'NO2': [(0, 40, 0, 50), (41, 80, 51, 100), (81, 180, 101, 200), (181, 280, 201, 300), (281, 400, 301, 400), (401, np.inf, 401, 500)],
    'CO': [(0.0, 1.0, 0, 50), (1.1, 2.0, 51, 100), (2.1, 10.0, 101, 200), (10.1, 17.0, 201, 300), (17.1, 34.0, 301, 400), (34.1, np.inf, 401, 500)], # in mg/m3
    'SO2': [(0, 40, 0, 50), (41, 80, 51, 100), (81, 380, 101, 200), (381, 800, 201, 300), (801, 1600, 301, 400), (1601, np.inf, 401, 500)],
    'O3': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 168, 101, 200), (169, 208, 201, 300), (209, 748, 301, 400), (749, np.inf, 401, 500)], # 8-hr average/max
    'NH3': [(0, 200, 0, 50), (201, 400, 51, 100), (401, 800, 101, 200), (801, 1200, 201, 300), (1201, 1800, 301, 400), (1801, np.inf, 401, 500)]
}

# 2. Daily data - pollutant averaging times setup
pollutant_averaging_times = {
    'PM2.5': {'window': '1D', 'min_periods': 1, 'method': 'mean'},
    'PM10': {'window': '1D', 'min_periods': 1, 'method': 'mean'},
    'NO2': {'window': '1D', 'min_periods': 1, 'method': 'mean'},
    'SO2': {'window': '1D', 'min_periods': 1, 'method': 'mean'},
    'NH3': {'window': '1D', 'min_periods': 1, 'method': 'mean'},
    'CO': {'window': '1D', 'min_periods': 1, 'method': 'max'},
    'O3': {'window': '1D', 'min_periods': 1, 'method': 'max'}
}

# 3. calculate_sub_index function
def calculate_sub_index(pollutant_name, concentration):
    if pd.isna(concentration): return np.nan
    concentration = max(0, concentration)
    breakpoints = cpcb_breakpoints.get(pollutant_name)
    if not breakpoints: return np.nan
    for bp_low, bp_hi, aqi_low, aqi_hi in breakpoints:
        if bp_low <= concentration <= bp_hi:
            if bp_hi == bp_low: return float(aqi_low)
            sub_index = ((aqi_hi - aqi_low) / (bp_hi - bp_low)) * (concentration - bp_low) + aqi_low
            return round(sub_index)
    if concentration > breakpoints[-1][1]: return 500
    return np.nan

# 4. calculate_aqi_for_dataframe_daily function - MODIFIED
def calculate_aqi_for_dataframe_daily(df_input, date_col='Date', city_col='City', # Keep city_col here
                                pm25_col='PM2.5', pm10_col='PM10',
                                no2_col='NO2', co_col='CO', so2_col='SO2',
                                o3_col='O3', nh3_col='NH3',
                                co_is_ppm=False):

    df_copy = df_input.copy()
    if not pd.api.types.is_datetime64_any_dtype(df_copy[date_col]):
        df_copy[date_col] = pd.to_datetime(df_copy[date_col], errors='coerce')
    df_copy = df_copy.set_index(date_col).sort_index()

    if co_is_ppm and co_col in df_copy.columns:
        df_copy[co_col] = df_copy[co_col] * 1.145

    processed_concentrations = {}
    aqi_pollutants_map = {
        'PM2.5': pm25_col, 'PM10': pm10_col, 'NO2': no2_col,
        'CO': co_col, 'SO2': so2_col, 'O3': o3_col, 'NH3': nh3_col
    }

    for pollutant, col_name in aqi_pollutants_map.items():
        if col_name in df_copy.columns:
            processed_concentrations[pollutant] = df_copy[col_name]
        else:
            processed_concentrations[pollutant] = pd.Series(np.nan, index=df_copy.index)

    aqi_results_df = pd.DataFrame(index=df_copy.index)
    aqi_results_df.index.name = date_col

    for pollutant, series in processed_concentrations.items():
        aqi_results_df[f'AQI_Sub_{pollutant}'] = series.apply(lambda x: calculate_sub_index(pollutant, x))

    aqi_overall_list = []
    aqi_category_list = []

    for index, row in aqi_results_df.iterrows():
        current_sub_indices = [
            row[f'AQI_Sub_{p}'] for p in cpcb_breakpoints.keys()
            if f'AQI_Sub_{p}' in row.index and pd.notna(row[f'AQI_Sub_{p}'])
        ]
        has_pm = pd.notna(row.get('AQI_Sub_PM2.5')) or pd.notna(row.get('AQI_Sub_PM10'))
        
        if len(current_sub_indices) >= 3 and has_pm:
            overall_aqi_val = max(current_sub_indices)
            aqi_overall_list.append(overall_aqi_val)
            category = (
                'Good' if 0 <= overall_aqi_val <= 50 else
                'Satisfactory' if 51 <= overall_aqi_val <= 100 else
                'Moderate' if 101 <= overall_aqi_val <= 200 else
                'Poor' if 201 <= overall_aqi_val <= 300 else
                'Very Poor' if 301 <= overall_aqi_val <= 400 else
                'Severe' if 401 <= overall_aqi_val <= 500 else
                np.nan
            )
            aqi_category_list.append(category)
        else:
            aqi_overall_list.append(np.nan)
            aqi_category_list.append(np.nan)

    aqi_results_df['AQI'] = aqi_overall_list
    aqi_results_df['AQI_Bucket'] = aqi_category_list

    # --- ADDITION HERE: Include the City column if it was in the input df_input ---
    if city_col in df_input.columns:
        # Before resetting index, add the City column back from the original df_input's index
        # This assumes df_input itself was indexed by Date (which it is internally after set_index)
        # We need to grab City from the original df_input's values corresponding to the current index
        aqi_results_df[city_col] = df_copy[city_col] # Directly add the Series from df_copy (which is indexed by Date)

    return aqi_results_df.reset_index()


# --- Core Logic to Calculate and Update AQI in your DataFrame (df) ---

# Assume 'df' is your main DataFrame and it has all pollutant columns filled.
# Ensure 'Date' column is datetime and 'City' column exists.

# 1. Create df_aqi_null: Select rows where 'AQI' is NaN.
df_aqi_null = df[df['AQI'].isna()].copy()

if len(df_aqi_null) == 0:
    print("No null AQI values found in 'df'. No calculation needed.")
else:
    print(f"Calculating AQI for {len(df_aqi_null)} rows with missing AQI values...")

    # 2. Calculate AQI for the subset df_aqi_null
    # Pass all relevant pollutant column names. Adjust co_is_ppm if necessary.
    # The 'city_col' parameter in the function ensures it's returned.
    calculated_aqi_for_nulls = calculate_aqi_for_dataframe_daily(
        df_input=df_aqi_null,
        date_col='Date',
        city_col='City', # Pass 'City' column name here so the function knows to handle it
        pm25_col='PM2.5', pm10_col='PM10',
        no2_col='NO2', co_col='CO', so2_col='SO2',
        o3_col='O3', nh3_col='NH3',
        co_is_ppm=False
    )

    # 3. Update the original 'df' with the newly calculated AQI values.
    # The merge ID approach is still the most robust.

    # Create a temporary unique ID for merging/updating
    df['_temp_id'] = df['City'].astype(str) + '_' + df['Date'].dt.strftime('%Y-%m-%d')
    calculated_aqi_for_nulls['_temp_id'] = calculated_aqi_for_nulls['City'].astype(str) + '_' + calculated_aqi_for_nulls['Date'].dt.strftime('%Y-%m-%d')

    # Set temporary index for efficient updating
    df_temp_indexed = df.set_index('_temp_id')
    calculated_aqi_indexed = calculated_aqi_for_nulls.set_index('_temp_id')

    # Update 'AQI' and 'AQI_Bucket' columns in df_temp_indexed
    df_temp_indexed.loc[calculated_aqi_indexed.index, 'AQI'] = calculated_aqi_indexed['AQI']
    df_temp_indexed.loc[calculated_aqi_indexed.index, 'AQI_Bucket'] = calculated_aqi_indexed['AQI_Bucket']

    # Assign back to original df and drop the temporary ID column
    df = df_temp_indexed.reset_index(drop=True)
    #df = df.drop(columns=['_temp_id'])

    print("AQI calculation and update complete.")
    print(f"Final NaNs in 'AQI' column: {df['AQI'].isnull().sum()}")
    print(f"Final NaNs in 'AQI_Bucket' column: {df['AQI_Bucket'].isnull().sum()}")
```
